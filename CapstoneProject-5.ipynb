## Procuring data from the website using read_html in pandas

```python
import pandas as pd
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
df = pd.read_html(url)[0]
df.head()
```

## Removing the missing data i.e the Not assigned Borough rows

```python
df0 = df[df.Borough!='Not assigned']
df0.reset_index(inplace=True,drop=True)
df0
```

## Changing the Not assigned Neighborhood values to respective Boroughs

```python
for row in df0.index:
    if(df0['Neighborhood'][row]=='Not assigned'):
        df0['Neighborhood'][row] = df0['Borough'][row]
        
```
